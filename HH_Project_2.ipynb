```python
# Устанавливаем соединение с базой данных
connection = psycopg2.connect(
    dbname=DBNAME,
    user=USER,
    password=PASSWORD,
    host=HOST,
    port=PORT
)

# Создаем курсор для выполнения SQL-запросов
cursor = connection.cursor()

# SQL-запрос для получения названия компании на третьем месте по алфавиту среди компаний
# у которых указано четыре сферы деятельности
query = """
SELECT name
FROM employers
WHERE id IN (
    SELECT employer_id
    FROM employers_industries
    GROUP BY employer_id
    HAVING COUNT(DISTINCT industry_id) = 4
)
ORDER BY name
LIMIT 1 OFFSET 2; -- Выбираем третью компанию из отсортированного списка
"""

# Выполняем SQL-запрос
cursor.execute(query)

# Получаем результат
result = cursor.fetchone()

# Закрываем курсор и соединение
cursor.close()
connection.close()

# Выводим результат на экран
if result:
    company_name = result[0]
    print(f"Название компании на третьем месте по алфавиту среди компаний с 4 сферами деятельности: {company_name}")
else:
    print("Компания не найдена.")
```

6. Напишите запрос, позволяющий вычислить, какую зарплату для DS в **среднем** указывают для каждого типа требуемого опыта (уникальное значение из поля *experience*).

При решении задачи примите во внимание следующее:
1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.
3. Если в расчётах участвует null, в результате он тоже даст null (посмотрите, что возвращает запрос select 1 + null). Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит null на значение, которое мы передадим. Например, посмотрите, что возвращает запрос `select 1 + coalesce(null, 0)`

Выясните, на какую зарплату в среднем может рассчитывать дата-сайентист с опытом работы от 3 до 6 лет. Результат округлите до целого числа.

```python
# Устанавливаем соединение с базой данных
connection = psycopg2.connect(
    dbname=DBNAME,
    user=USER,
    password=PASSWORD,
    host=HOST,
    port=PORT
)

# Создаем курсор для выполнения SQL-запросов
cursor = connection.cursor()

# SQL-запрос для вычисления средней зарплаты для DS с опытом от 3 до 6 лет
query = """
SELECT
    experience,
    ROUND(AVG(
        CASE
            WHEN salary_from IS NOT NULL AND salary_to IS NOT NULL THEN (salary_from + salary_to) / 2
            WHEN salary_from IS NOT NULL THEN salary_from
            WHEN salary_to IS NOT NULL THEN salary_to
            ELSE null
        END
    )) AS average_salary
FROM vacancies
WHERE
    (LOWER(name) LIKE '%data scientist%'
    OR LOWER(name) LIKE '%data science%'
    OR LOWER(name) LIKE '%исследователь данных%'
    OR LOWER(name) LIKE '%machine learning%'
    OR (LOWER(name) LIKE '%машинн%' AND LOWER(name) LIKE '%обучен%')
    OR LOWER(name) LIKE '%ml%')
    AND experience ~ '^\d+'
    AND CAST(experience AS INTEGER) BETWEEN 3 AND 6
GROUP BY experience;
"""

# Выполняем SQL-запрос
cursor.execute(query)

# Получаем результат
results = cursor.fetchall()

# Закрываем курсор и соединение
cursor.close()
connection.close()

# Выводим результат на экран
for row in results:
    experience, average_salary = row
    print(f"Опыт работы от {experience} лет: Средняя зарплата - {average_salary}")
```
